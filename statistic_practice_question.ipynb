````python
Q= Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
After generating the list, find the following:



(i) Write a Python function to calculate the mean of a given list of numbers.
Create a function to find the median of a list of numbers.
import random

# Generate the list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to calculate the median
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:  # Even number of elements
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:  # Odd number of elements
        return sorted_numbers[mid]

# Calculate mean and median
mean_value = calculate_mean(int_list)
median_value = calculate_median(int_list)

# Output the results
print("Generated List:", int_list)
print("Mean:", mean_value)
print("Median:", median_value)

  (ii) Develop a program to compute the mode of a list of integers.
  import random
from collections import Counter

# Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to compute the mode
def compute_mode(data):
    # Count the frequency of each element
    freq = Counter(data)
    # Find the maximum frequency
    max_freq = max(freq.values())
    # Find all elements that have the maximum frequency
    mode = [num for num, count in freq.items() if count == max_freq]
    return mod

# Compute the mode of the generated list
mode = compute_mode(int_list)
print("Mode of the list:", mode)

  (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
  import random

# Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the weighted mean
def weighted_mean(values, weights):
    """
    Calculate the weighted mean of a list of values with corresponding weights.

    Parameters:
    values (list): A list of values.
    weights (list): A list of weights corresponding to the values.

    Returns:
    float: The weighted mean of the values.
    """
    if len(values) != len(weights):
        raise ValueError("The length of values and weights must be the same.")

    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)

    if total_weight == 0:
        raise ValueError("The sum of weights must not be zero.")

    return weighted_sum / total_weight

# Example usage
values = int_list
weights = [1] * len(int_list)  # Example weights, all equal to 1
weighted_mean_result = weighted_mean(values, weights)
print("Weighted Mean:", weighted_mean_result)

  (iv) Write a Python function to find the geometric mean of a list of positive numbers.
  import random
import math
from functools import reduce

# Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the geometric mean
def geometric_mean(values):
    """
    Calculate the geometric mean of a list of positive numbers.

    Parameters:
    values (list): A list of positive numbers.

    Returns:
    float: The geometric mean of the numbers.
    """
    if not all(v > 0 for v in values):
        raise ValueError("All numbers must be positive.")

    product = reduce(lambda x, y: x * y, values)
    n = len(values)

    return product ** (1/n)  # Alternatively, you can use math.pow(product, 1/n)

# Calculate the geometric mean of the generated list
geo_mean_result = geometric_mean(int_list)
print("Geometric Mean:", geo_mean_result)

  (v) Create a program to calculate the harmonic mean of a list of values.
  import random

# Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the harmonic mean
def harmonic_mean(values):
    """
    Calculate the harmonic mean of a list of positive numbers.

    Parameters:
    values (list): A list of positive numbers.

    Returns:
    float: The harmonic mean of the numbers.
    """
    if not all(v > 0 for v in values):
        raise ValueError("All numbers must be positive.")

    n = len(values)
    reciprocal_sum = sum(1 / v for v in values)

    return n / reciprocal_sum  # Harmonic mean formula

# Calculate the harmonic mean of the generated list
harmonic_mean_result = harmonic_mean(int_list)
print("Harmonic Mean:", harmonic_mean_result)

  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
  import random

# Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the harmonic mean
def harmonic_mean(values):
    """
    Calculate the harmonic mean of a list of positive numbers.

    Parameters:
    values (list): A list of positive numbers.

    Returns:
    float: The harmonic mean of the numbers.
    """
    if not all(v > 0 for v in values):
        raise ValueError("All numbers must be positive.")

    n = len(values)
    reciprocal_sum = sum(1 / v for v in values)

    return n / reciprocal_sum  # Harmonic mean formula

# Calculate the harmonic mean of the generated list
harmonic_mean_result = harmonic_mean(int_list)
print("Harmonic Mean:", harmonic_mean_result)

(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers.
import random

# Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the trimmed mean
def trimmed_mean(values, trim_percent):
    """
    Calculate the trimmed mean of a list of numbers, excluding a certain percentage of outliers.

    Parameters:
    values (list): A list of numerical values.
    trim_percent (float): The percentage of values to trim from each end of the sorted list.

    Returns:
    float: The trimmed mean of the numbers.
    """
    if not 0 <= trim_percent < 1:
        raise ValueError("trim_percent must be between 0 and 1.")

    # Sort the values
    sorted_values = sorted(values)

    # Calculate number of elements to trim
    n = len(sorted_values)
    trim_count = int(n * trim_percent)

    # Trim the sorted values
    trimmed_values = sorted_values[trim_count:n-trim_count]

    if len(trimmed_values) == 0:
        raise ValueError("Not enough data to calculate trimmed mean after trimming.")

    # Calculate and return the mean of the trimmed values
    return sum(trimmed_values) / len(trimmed_values)

# Example usage with a trim percentage of 10%
trim_percent = 0.10
trimmed_mean_result = trimmed_mean(int_list, trim_percent)
print("Trimmed Mean (10% trimmed):", trimmed_mean_result)

 2. Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
After generating the list, find the following:


  (i) Compare the given list of visualization for the given data:



    1. Frequency & Gaussian distribution

    2. Frequency smoothened KDE plot

    3. Gaussian distribution & smoothened KDE plot


import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Convert to a NumPy array for easier handling
data = np.array(int_list2)

# Step 2: Set up the figure for plotting
plt.figure(figsize=(15, 10))

# 1. Frequency Histogram & Gaussian Distribution
plt.subplot(2, 2, 1)
sns.histplot(data, bins=30, kde=False, stat="density", color='blue', edgecolor='black', alpha=0.7)
mean = np.mean(data)
std_dev = np.std(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'r', linewidth=2)
plt.title('Frequency Histogram with Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()

# 2. Frequency Smoothened KDE Plot
plt.subplot(2, 2, 2)
sns.kdeplot(data, fill=True, color='blue', alpha=0.5)
plt.title('Smoothened KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()

# 3. Gaussian Distribution & Smoothened KDE Plot
plt.subplot(2, 2, (3, 4))
sns.kdeplot(data, fill=True, color='blue', alpha=0.5, label='KDE')
plt.plot(x, p, 'r', linewidth=2, label='Gaussian Distribution')
plt.title('Gaussian Distribution & Smoothened KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()

# Show the plots
plt.tight_layout()
plt.show()
  (ii) Write a Python function to calculate the range of a given list of numbers.
  import random

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the range of a list of numbers
def calculate_range(values):
    """
    Calculate the range of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The range of the numbers.
    """
    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    min_value = min(values)
    max_value = max(values)

    return max_value - min_value  # Calculate range

# Step 3: Calculate the range of the generated list
range_result = calculate_range(int_list2)
print("Range of the list:", range_result)

  (iii) Create a program to find the variance and standard deviation of a list of numbers.
  import random
import math

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the mean
def calculate_mean(values):
    """
    Calculate the mean of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The mean of the numbers.
    """
    return sum(values) / len(values)

# Step 3: Function to calculate the variance
def calculate_variance(values):
    """
    Calculate the variance of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The variance of the numbers.
    """
    mean = calculate_mean(values)
    return sum((x - mean) ** 2 for x in values) / len(values)

# Step 4: Function to calculate the standard deviation
def calculate_standard_deviation(values):
    """
    Calculate the standard deviation of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The standard deviation of the numbers.
    """
    variance = calculate_variance(values)
    return math.sqrt(variance)

# Step 5: Calculate the variance and standard deviation of the generated list
variance_result = calculate_variance(int_list2)
standard_deviation_result = calculate_standard_deviation(int_list2)

# Step 6: Print the results
print("Variance of the list:", variance_result)
print("Standard Deviation of the list:", standard_deviation_result)
  (iv) Implement a function to compute the interquartile range (IQR) of a list of values.
  import random
import numpy as np

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the interquartile range (IQR)
def compute_iqr(values):
    """
    Calculate the interquartile range (IQR) of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The interquartile range of the numbers.
    """
    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    # Sort the values
    sorted_values = sorted(values)

    # Calculate the first (Q1) and third (Q3) quartiles
    Q1 = np.percentile(sorted_values, 25)
    Q3 = np.percentile(sorted_values, 75)

    # Calculate the IQR
    IQR = Q3 - Q1
    return IQR

# Step 3: Calculate the IQR of the generated list
iqr_result = compute_iqr(int_list2)

# Step 4: Print the result
print("Interquartile Range (IQR) of the list:", iqr_result)
  (v) Build a program to calculate the coefficient of variation for a dataset.
  import random
import math

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the mean
def calculate_mean(values):
    """
    Calculate the mean of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The mean of the numbers.
    """
    return sum(values) / len(values)

# Step 3: Function to calculate the standard deviation
def calculate_standard_deviation(values):
    """
    Calculate the standard deviation of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The standard deviation of the numbers.
    """
    mean = calculate_mean(values)
    variance = sum((x - mean) ** 2 for x in values) / len(values)
    return math.sqrt(variance)

# Step 4: Function to calculate the coefficient of variation
def calculate_coefficient_of_variation(values):
    """
    Calculate the coefficient of variation (CV) of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The coefficient of variation as a percentage.
    """
    mean = calculate_mean(values)
    std_dev = calculate_standard_deviation(values)

    if mean == 0:
        raise ValueError("Mean cannot be zero, cannot calculate CV.")

    cv = (std_dev / mean) * 100  # Coefficient of Variation as a percentage
    return cv

# Step 5: Calculate the coefficient of variation of the generated list
cv_result = calculate_coefficient_of_variation(int_list2)

# Step 6: Print the result
print("Coefficient of Variation (CV) of the list: {:.2f}%".format(cv_result))
  (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.
  import random

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the mean
def calculate_mean(values):
    """
    Calculate the mean of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The mean of the numbers.
    """
    return sum(values) / len(values)

# Step 3: Function to calculate the mean absolute deviation (MAD)
def calculate_mean_absolute_deviation(values):
    """
    Calculate the Mean Absolute Deviation (MAD) of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The mean absolute deviation of the numbers.
    """
    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    mean = calculate_mean(values)
    absolute_deviations = [abs(x - mean) for x in values]  # Calculate absolute deviations
    mad = calculate_mean(absolute_deviations)  # Calculate the mean of the absolute deviations
    return mad

# Step 4: Calculate the Mean Absolute Deviation of the generated list
mad_result = calculate_mean_absolute_deviation(int_list2)

# Step 5: Print the result
print("Mean Absolute Deviation (MAD) of the list:", mad_result)
  (vii) Create a program to calculate the quartile deviation of a list of values.
  import random
import numpy as np

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the quartile deviation
def calculate_quartile_deviation(values):
    """
    Calculate the quartile deviation of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The quartile deviation of the numbers.
    """
    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    # Calculate Q1 and Q3 using numpy
    Q1 = np.percentile(values, 25)
    Q3 = np.percentile(values, 75)

    # Calculate the quartile deviation
    quartile_deviation = (Q3 - Q1) / 2
    return quartile_deviation

# Step 3: Calculate the quartile deviation of the generated list
quartile_deviation_result = calculate_quartile_deviation(int_list2)

# Step 4: Print the result
print("Quartile Deviation of the list:", quartile_deviation_result)
  (viii) Implement a function to find the range-based coefficient of dispersion for a dataset.
  import random

# Step 1: Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Step 2: Function to calculate the mean
def calculate_mean(values):
    """
    Calculate the mean of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The mean of the numbers.
    """
    return sum(values) / len(values)

# Step 3: Function to calculate the range
def calculate_range(values):
    """
    Calculate the range of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The range of the numbers.
    """
    return max(values) - min(values)

# Step 4: Function to calculate the range-based coefficient of dispersion
def calculate_range_based_coefficient_of_dispersion(values):
    """
    Calculate the range-based coefficient of dispersion of a list of numbers.

    Parameters:
    values (list): A list of numerical values.

    Returns:
    float: The range-based coefficient of dispersion as a percentage.
    """
    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    mean = calculate_mean(values)
    if mean == 0:
        raise ValueError("Mean cannot be zero, cannot calculate coefficient of dispersion.")

    range_value = calculate_range(values)

    coefficient_of_dispersion = (range_value / mean) * 100  # Coefficient of Dispersion as a percentage
    return coefficient_of_dispersion

# Step 5: Calculate the range-based coefficient of dispersion of the generated list
dispersion_result = calculate_range_based_coefficient_of_dispersion(int_list2)

# Step 6: Print the result
print("Range-based Coefficient of Dispersion of the list: {:.2f}%".format(dispersion_result))

3)  Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize the discrete random variable with outcomes and their probabilities.

        Parameters:
        outcomes (list): A list of possible outcomes.
        probabilities (list): A list of probabilities corresponding to each outcome.
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("Outcomes and probabilities must be of the same length.")
        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if not abs(sum(probabilities) - 1) < 1e-10:  # Check if probabilities sum to 1
            raise ValueError("Probabilities must sum to 1.")

        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value of the discrete random variable.

        Returns:
        float: The expected value (mean) of the random variable.
        """
        return sum(outcome * prob for outcome, prob in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the discrete random variable.

        Returns:
        float: The variance of the random variable.
        """
        mean = self.expected_value()
        return sum(prob * (outcome - mean) ** 2 for outcome, prob in zip(self.outcomes, self.probabilities))

# Example usage
if __name__ == "__main__":
    # Define outcomes and their probabilities
    outcomes = [0, 1, 2, 3, 4]
    probabilities = [0.1, 0.2, 0.4, 0.2, 0.1]  # Must sum to 1

    # Create a DiscreteRandomVariable instance
    random_variable = DiscreteRandomVariable(outcomes, probabilities)

    # Calculate expected value and variance
    expected_val = random_variable.expected_value()
    variance_val = random_variable.variance()

    # Print the results
    print("Expected Value:", expected_val)
    print("Variance:", variance_val)

    4) Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes.
class FairDie:
    def __init__(self):
        self.outcomes = [1, 2, 3, 4, 5, 6]
        self.probabilities = [1/6] * 6  # Equal probability for a fair die

    def expected_value(self):
        """
        Calculate the expected value of rolling the die.

        Returns:
        float: The expected value (mean) of the die outcomes.
        """
        return sum(outcome * prob for outcome, prob in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the outcomes when rolling the die.

        Returns:
        float: The variance of the die outcomes.
        """
        mean = self.expected_value()
        return sum(prob * (outcome - mean) ** 2 for outcome, prob in zip(self.outcomes, self.probabilities))

# Example usage
if __name__ == "__main__":
    die = FairDie()

    # Calculate expected value and variance
    expected_val = die.expected_value()
    variance_val = die.variance()

    # Print the results
    print("Expected Value:", expected_val)
    print("Variance:", variance_val)

    5) Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.
import numpy as np

def generate_random_samples(distribution, **kwargs):
    """
    Generate random samples from a specified probability distribution
    and calculate their mean and variance.

    Parameters:
    distribution (str): The type of distribution ('binomial' or 'poisson').
    kwargs: Additional parameters for the distribution.

    Returns:
    tuple: A tuple containing the generated samples, their mean, and variance.
    """
    if distribution == 'binomial':
        n = kwargs.get('n')  # Number of trials
        p = kwargs.get('p')  # Probability of success
        size = kwargs.get('size', 1000)  # Default number of samples
        samples = np.random.binomial(n, p, size)

    elif distribution == 'poisson':
        lam = kwargs.get('lam')  # Lambda (mean number of events)
        size = kwargs.get('size', 1000)  # Default number of samples
        samples = np.random.poisson(lam, size)

    else:
        raise ValueError("Unsupported distribution type. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)

    return samples, mean, variance

# Example usage
if __name__ == "__main__":
    # Generate samples from a binomial distribution
    binomial_samples, binomial_mean, binomial_variance = generate_random_samples(
        distribution='binomial',
        n=10,    # Number of trials
        p=0.5,   # Probability of success
        size=1000  # Number of samples
    )

    print("Binomial Distribution:")
    print("Mean:", binomial_mean)
    print("Variance:", binomial_variance)

    # Generate samples from a Poisson distribution
    poisson_samples, poisson_mean, poisson_variance = generate_random_samples(
        distribution='poisson',
        lam=5,     # Lambda (mean number of events)
        size=1000   # Number of samples
    )

    print("\nPoisson Distribution:")
    print("Mean:", poisson_mean)
    print("Variance:", poisson_variance)

    6) Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
the mean, variance, and standard deviation of the samples.
import numpy as np

def generate_normal_samples(mean=0, std_dev=1, size=1000):
    """
    Generate random samples from a Gaussian (normal) distribution
    and compute their mean, variance, and standard deviation.

    Parameters:
    mean (float): The mean of the normal distribution.
    std_dev (float): The standard deviation of the normal distribution.
    size (int): The number of samples to generate.

    Returns:
    tuple: A tuple containing the generated samples, their mean, variance, and standard deviation.
    """
    # Generate random samples from a normal distribution
    samples = np.random.normal(mean, std_dev, size)

    # Calculate statistics
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)

    return samples, sample_mean, sample_variance, sample_std_dev

# Example usage
if __name__ == "__main__":
    # Parameters for the normal distribution
    mean = 0       # Mean of the distribution
    std_dev = 1    # Standard deviation of the distribution
    size = 1000    # Number of samples to generate

    # Generate samples and compute statistics
    samples, computed_mean, computed_variance, computed_std_dev = generate_normal_samples(mean, std_dev, size)

    # Print the results
    print("Generated Samples:", samples[:10], "...")  # Print first 10 samples
    print("Mean of samples:", computed_mean)
    print("Variance of samples:", computed_variance)
    print("Standard Deviation of samples:", computed_std_dev)

    7) Use seaborn library to load tips dataset. Find the following from the dataset for the columns total_bill
and tip`:



  (i) Write a Python function that calculates their skewness.
  import seaborn as sns
import pandas as pd
from scipy.stats import skew

def calculate_skewness(data, column):
    """
    Calculate the skewness of a specified column in the dataset.

    Parameters:
    data (DataFrame): The dataset containing the column.
    column (str): The name of the column to calculate skewness for.

    Returns:
    float: The skewness of the specified column.
    """
    return skew(data[column])

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate skewness for total_bill and tip
total_bill_skewness = calculate_skewness(tips, 'total_bill')
tip_skewness = calculate_skewness(tips, 'tip')

# Print the results
print("Skewness of total_bill:", total_bill_skewness)
print("Skewness of tip:", tip_skewness)
(ii) Create a program that determines whether the columns exhibit positive skewness, negative skewness, or is
approximately symmetric.
import seaborn as sns
import pandas as pd
from scipy.stats import skew

def calculate_skewness(data, column):
    """
    Calculate the skewness of a specified column in the dataset.

    Parameters:
    data (DataFrame): The dataset containing the column.
    column (str): The name of the column to calculate skewness for.

    Returns:
    float: The skewness of the specified column.
    """
    return skew(data[column])

def classify_skewness(skewness_value):
    """
    Classify the skewness as positive, negative, or symmetric.

    Parameters:
    skewness_value (float): The skewness value to classify.

    Returns:
    str: A description of the skewness classification.
    """
    if skewness_value > 0:
        return "Positive Skewness"
    elif skewness_value < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate skewness for total_bill and tip
total_bill_skewness = calculate_skewness(tips, 'total_bill')
tip_skewness = calculate_skewness(tips, 'tip')

# Classify the skewness
total_bill_skewness_classification = classify_skewness(total_bill_skewness)
tip_skewness_classification = classify_skewness(tip_skewness)

# Print the results
print("Skewness of total_bill:", total_bill_skewness, "->", total_bill_skewness_classification)
print("Skewness of tip:", tip_skewness, "->", tip_skewness_classification)
  (iii) Write a function that calculates the covariance between two columns.
  import seaborn as sns
import pandas as pd

def calculate_covariance(data, column1, column2):
    """
    Calculate the covariance between two specified columns in the dataset.

    Parameters:
    data (DataFrame): The dataset containing the columns.
    column1 (str): The name of the first column.
    column2 (str): The name of the second column.

    Returns:
    float: The covariance between the two columns.
    """
    return data[column1].cov(data[column2])

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate covariance between 'total_bill' and 'tip'
covariance_total_bill_tip = calculate_covariance(tips, 'total_bill', 'tip')

# Print the result
print("Covariance between total_bill and tip:", covariance_total_bill_tip)
  (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.
  import seaborn as sns
import pandas as pd

def calculate_pearson_correlation(data, column1, column2):
    """
    Calculate the Pearson correlation coefficient between two specified columns in the dataset.

    Parameters:
    data (DataFrame): The dataset containing the columns.
    column1 (str): The name of the first column.
    column2 (str): The name of the second column.

    Returns:
    float: The Pearson correlation coefficient between the two columns.
    """
    return data[column1].corr(data[column2])

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate Pearson correlation coefficient between 'total_bill' and 'tip'
pearson_correlation_total_bill_tip = calculate_pearson_correlation(tips, 'total_bill', 'tip')

# Print the result
print("Pearson correlation coefficient between total_bill and tip:", pearson_correlation_total_bill_tip)
 (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using
scatter plots.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a scatter plot to visualize the correlation between total_bill and tip
plt.figure(figsize=(10, 6))  # Set the figure size
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', palette='deep')

# Add title and labels
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# Show the plot
plt.legend(title='Day/Time')
plt.grid(True)  # Optional: add a grid for easier reading
plt.show()

8) Write a Python function to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.
import numpy as np

def normal_pdf(x, mean=0, std_dev=1):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    x (float): The value at which to evaluate the PDF.
    mean (float): The mean of the normal distribution (default = 0).
    std_dev (float): The standard deviation of the normal distribution (default = 1).

    Returns:
    float: The value of the PDF at the specified x.
    """
    # Calculate the PDF using the normal distribution formula
    coefficient = 1 / (std_dev * np.sqrt(2 * np.pi))
    exponent = np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    return coefficient * exponent

# Example usage
if __name__ == "__main__":
    # Parameters for the normal distribution
    mean = 0
    std_dev = 1

    # Calculate PDF for a range of x values
    x_values = np.linspace(-4, 4, 100)  # Generate 100 points from -4 to 4
    pdf_values = [normal_pdf(x, mean, std_dev) for x in x_values]

    # Print some example PDF values
    for x, pdf in zip(x_values, pdf_values):
        print(f"PDF at x={x:.2f}: {pdf:.5f}")

9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.
import math

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    Parameters:
    x (float): The value at which to evaluate the CDF.
    rate (float): The rate parameter of the exponential distribution.

    Returns:
    float: The value of the CDF at the specified x.
    """
    if x < 0:
        return 0
    else:
        return 1 - math.exp(-rate * x)

# Example usage
if __name__ == "__main__":
    # Parameters for the exponential distribution
    rate = 0.5  # Example rate parameter
    x_values = [0, 1, 2, 3, 4]  # Example x values

    # Calculate CDF for the given x values
    cdf_values = [exponential_cdf(x, rate) for x in x_values]

    # Print the results
    for x, cdf in zip(x_values, cdf_values):
        print(f"CDF at x={x}: {cdf:.5f}")

10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.
import math

def poisson_pmf(k, lambd):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.

    Parameters:
    k (int): The number of occurrences.
    lambd (float): The expected number of occurrences in the given interval.

    Returns:
    float: The probability of k occurrences.
    """
    if k < 0 or lambd <= 0:
        raise ValueError("k must be non-negative and lambd must be positive.")

    pmf = (math.exp(-lambd) * (lambd ** k)) / math.factorial(k)
    return pmf

# Example usage
if __name__ == "__main__":
    lambd = 3  # Example lambda value
    k_values = [0, 1, 2, 3, 4, 5]  # Example k values

    # Calculate PMF for the given k values
    pmf_values = [poisson_pmf(k, lambd) for k in k_values]

    # Print the results
    for k, pmf in zip(k_values, pmf_values):
        print(f"PMF at k={k}: {pmf:.5f}")

11) A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.


To generate the data use the following command:

```python

import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors

new_layout = np.array([1] * 70 + [0] * 930)

  ```Apply z-test to find which layout is successful.
  import numpy as np
import scipy.stats as stats

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930  )  # 70 purchases out of 1000 visitors

# Calculate the number of successes (purchases) and number of trials (visitors)
success_old = np.sum(old_layout)
n_old = len(old_layout)

success_new = np.sum(new_layout)
n_new = len(new_layout)

# Calculate conversion rates
p_old = success_old / n_old
p_new = success_new / n_new

# Calculate the combined proportion
p_combined = (success_old + success_new) / (n_old + n_new)

# Calculate the z-score
z = (p_new - p_old) / np.sqrt(p_combined * (1 - p_combined) * (1/n_old + 1/n_new))

# Calculate the p-value for a one-tailed test (since we want to see if new layout is better)
p_value = 1 - stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Old Layout Conversion Rate: {p_old:.4f}")
print(f"New Layout Conversion Rate: {p_new:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new layout leads to a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")

12) A tutoring service claims that its program improves students' exam scores. A sample of students who
participated in the program was taken, and their scores before and after the program were recorded.


Use the below code to generate samples of respective arrays of marks:

```python

before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.
import numpy as np
import scipy.stats as stats

# Generate the data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences
differences = after_program - before_program

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = 1 - stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program significantly improves exam scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in exam scores.")

13) A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They
conduct a study and record blood pressure measurements before and after administering the drug.


Use the below code to generate samples of respective arrays of blood pressure:


```python

before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])

after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

  ```


Implement z-test to find if the drug really works or not.
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

14) A customer service department claims that their average response time is less than 5 minutes. A sample
of recent customer interactions was taken, and the response times were recorded.


Implement the below code to generate the array of response time:

```python

response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

```

Implement z-test to find the claims made by customer service department are tru or false.
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

  15) A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.


Use the following data:

```python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

    16) A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in
reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the tstatistic and p-value for the treatment effect.


Use the following data of cholestrol level:

```python

existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]

new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

    17) A school district introduces an educational intervention program to improve math scores. Write a Python
function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
determine if the intervention had a significant impact.


Use the following data of test score:


  ```python

  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]

  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
  import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

    18) An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
significant difference between the average salaries of male and female employees.


Use the below code to generate synthetic data:


```python

# Generate synthetic salary data for male and female employees

np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)

female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

    19) A manufacturer produces two different versions of a product and wants to compare their quality scores.
Create a Python function to analyze quality assessment data, calculate the t-statistic, and decide
whether there's a significant difference in quality between the two versions.


Use the following data:


```python

version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]

version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

     20. A restaurant chain collects customer satisfaction scores for two different branches. Write a program to
analyze the scores, calculate the t-statistic, and determine if there's a statistically significant difference in
customer satisfaction between the branches.


Use the below data of scores:

  ```python

branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]

branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]
import numpy as np
import scipy.stats as stats

# Generate the data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = after_drug - before_drug

# Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-score
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(z)  # One-tailed test

# Output the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug significantly reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

    21) A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Data generation
np.random.seed(0)

# Define age groups and voter preferences
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Display results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

# Conclusion
alpha = 0.05
if p < alpha:
    print("\nConclusion: There is a significant association between age groups and voter preferences.")
else:
    print("\nConclusion: There is no significant association between age groups and voter preferences.")

22. A company conducted a customer satisfaction survey to determine if there is a significant relationship
between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are
located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a ChiSquare test to determine if there is a significant relationship between product satisfaction levels and
customer regions.


Sample data:

```python

#Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)

data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])
import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Contingency table
# Rows: Product satisfaction levels (Satisfied, Neutral, Dissatisfied)
# Columns: Customer regions (East, West, North, South)
data = np.array([[50, 30, 40, 20],
                 [30, 40, 30, 50],
                 [20, 30, 40, 30]])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

# Display results
print("Contingency Table (Observed):")
print(data)
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

# Conclusion
alpha = 0.05
if p < alpha:
    print("\nConclusion: There is a significant relationship between product satisfaction levels and customer regions.")
else:
    print("\nConclusion: There is no significant relationship between product satisfaction levels and customer regions.")

 23. A company implemented an employee training program to improve job performance (Effective, Neutral,
Ineffective). After the training, they collected data from a sample of employees and classified them based
on their job performance before and after the training. Perform a Chi-Square test to determine if there is a
significant difference between job performance levels before and after the training.


Sample data:

```python

# Sample data: Job performance levels before (rows) and after (columns) training

data = np.array([[50, 30, 20], [30, 40, 30], [20, 30, 40]])
import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Contingency table
# Rows: Job performance levels before training (Effective, Neutral, Ineffective)
# Columns: Job performance levels after training (Effective, Neutral, Ineffective)
data = np.array([[50, 30, 20],
                 [30, 40, 30],
                 [20, 30, 40]])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

# Display results
print("Contingency Table (Observed):")
print(data)
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

# Conclusion
alpha = 0.05
if p < alpha:
    print("\nConclusion: There is a significant difference in job performance levels before and after the training.")
else:
    print("\nConclusion: There is no significant difference in job performance levels before and after the training.")

 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The
company wants to determine if there is a significant difference in customer satisfaction scores among the
three product versions. They conducted a survey and collected customer satisfaction scores for each
version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
difference in customer satisfaction scores.


  Use the following data:

  ```python

  # Sample data: Customer satisfaction scores for each product version

  standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]

  premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

  deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
  from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform one-way ANOVA test
f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

# Display results
print("ANOVA Test Results:")
print(f"F-Statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: There is a significant difference in customer satisfaction scores among the three product versions.")
else:
    print("\nConclusion: There is no significant difference in customer satisfaction scores among the three product versions.")
````
